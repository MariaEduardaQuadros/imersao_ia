```python
# Check data types
print(dados['rating'].dtype)

# Convert 'rating' column to numeric
dados['rating'] = pd.to_numeric(dados['rating'], errors='coerce')

# Check if conversion was successful
print(dados['rating'].dtype)
```

```python
# Convert 'rating' column to numeric
dados['rating'] = pd.to_numeric(dados['rating'], errors='coerce')

# Drop rows with NaN values in the 'rating' column
dados.dropna(subset=['rating'], inplace=True)

# Check data types again
print(dados['rating'].dtype)
```

```python

# Now perform your analysis
ratings_above_4_5 = dados[dados['rating'] > 4.5]
rating_counts = ratings_above_4_5['rating'].value_counts().sort_index()

plt.figure(figsize=(10,6))
sns.barplot(x=rating_counts.index, y=rating_counts.values)
plt.title('Contagem de Avaliação acima de 4.5')
plt.xlabel('Avaliação')
plt.ylabel('Contagem')
plt.show()
```

```python
faixa1 = dados[dados['actual_price'] <= 50]
faixa2 = dados[dados['actual_price'] > 50 & (dados['actual_price'] <= 50)]
faixa3 = dados[dados['actual_price'] > 500]

contagem_faixa1 = len(faixa1)
contagem_faixa2 = len(faixa2)
contagem_faixa3 = len(faixa3)

contagens = [contagem_faixa1, contagem_faixa2, contagem_faixa3]
faixas = ['Até 50 ₹', 'Entre 50₹ e 500₹', 'Acima de 500₹']
explode = (0.1,0,0)

plt.figure(figsize=(8, 8))
plt.pie(contagens, labels=faixas, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Distribuição de profutos por faixa de valor')
plt.axis('equal')
plt.show()
```
